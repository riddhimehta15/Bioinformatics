```python
# =========================
# Cell 0: Imports & config
# =========================

import os
from collections import Counter

import numpy as np
import pandas as pd

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    matthews_corrcoef,
    roc_auc_score,
)

# Paths (adjust if needed)
POS_FASTA = "Positive_Final.fasta"   # positive = RBP
NEG_FASTA = "Negative_Final.fasta"   # negative = non-RBP

# Amino acid definitions
AA_ORDER = list("ACDEFGHIKLMNPQRSTVWY")
AA_INDEX = {aa: i for i, aa in enumerate(AA_ORDER)}

RANDOM_STATE = 42
```

```python
# =========================
# Cell 1: FASTA utilities
# =========================

def read_fasta(path):
    """
    Minimal FASTA reader.
    Returns dict: {sequence_id: sequence_string}
    """
    seqs = {}
    header = None
    parts = []
    with open(path, "r") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            if line.startswith(">"):
                if header is not None:
                    seqs[header] = "".join(parts).replace(" ", "").upper()
                header = line[1:].split()[0]   # take first token after '>'
                parts = []
            else:
                parts.append(line)
    if header is not None:
        seqs[header] = "".join(parts).replace(" ", "").upper()
    return seqs


# Load your dataset
pos_seqs = read_fasta(POS_FASTA)
neg_seqs = read_fasta(NEG_FASTA)

print(f"Loaded {len(pos_seqs)} positive (RBP) sequences")
print(f"Loaded {len(neg_seqs)} negative (non-RBP) sequences")

def build_labeled_df(pos_dict, neg_dict):
    rows = []
    for sid, seq in pos_dict.items():
        rows.append({"id": sid, "sequence": seq, "label": 1})
    for sid, seq in neg_dict.items():
        rows.append({"id": sid, "sequence": seq, "label": 0})
    return pd.DataFrame(rows)

df = build_labeled_df(pos_seqs, neg_seqs)

print(df.head())
print("\nLabel counts:")
print(df["label"].value_counts())
print("\nSequence length stats:")
print(df["sequence"].str.len().describe())
```

```python
# =========================
# Cell 2: AAC features
# =========================

def compute_aac(seq, aa_order=AA_ORDER):
    """
    Amino Acid Composition (AAC):
    p_i = l_i / L, where l_i is count of AA i and L is sequence length.
    Returns a len(aa_order)-dim numpy array.
    """
    counts = np.zeros(len(aa_order), dtype=float)
    L = 0
    for aa in seq:
        if aa in AA_INDEX:
            counts[AA_INDEX[aa]] += 1
            L += 1
    if L > 0:
        counts /= L
    return counts


# Quick sanity check
test_aac = compute_aac(df["sequence"].iloc[0])
print("AAC length:", len(test_aac))
print("AAC sum:", test_aac.sum())
```

```python
# =========================
# Cell 3: Dipeptide features
# =========================

# All possible dipeptides in fixed order: AA, AC, AD, ..., YY
DIPEPTIDES = [a + b for a in AA_ORDER for b in AA_ORDER]
DIPEP_INDEX = {dp: i for i, dp in enumerate(DIPEPTIDES)}

def compute_dipeptide_composition(seq):
    """
    Dipeptide composition:
    frequency of each possible pair (AA, AC, ..., YY) normalized by (L-1).
    Returns a 400-dim vector in DIPEPTIDES order.
    """
    L = len(seq)
    counts = np.zeros(len(DIPEPTIDES), dtype=float)
    if L < 2:
        return counts  # no dipeptides possible

    total = 0
    for i in range(L - 1):
        a1, a2 = seq[i], seq[i+1]
        if a1 in AA_INDEX and a2 in AA_INDEX:
            dp = a1 + a2
            idx = DIPEP_INDEX[dp]
            counts[idx] += 1
            total += 1

    if total > 0:
        counts /= total
    return counts


# Sanity check
test_dipep = compute_dipeptide_composition(df["sequence"].iloc[0])
print("Dipeptide vector length:", len(test_dipep))
print("Dipeptide sum:", test_dipep.sum())
```

```python
# =========================
# Cell A1: 3-mer (tri-peptide) features
# =========================

from itertools import product

# All possible 3-mers over 20 AA: 20^3 = 8000
TRIPEPTIDES = ["".join(p) for p in product(AA_ORDER, repeat=3)]
TRIPEP_INDEX = {tp: i for i, tp in enumerate(TRIPEPTIDES)}

def compute_tripeptide_composition(seq):
    """
    3-mer composition:
    frequency of each possible triplet (AAA, AAC, ..., YYY),
    normalized by (L-2).
    Returns an 8000-dim vector in TRIPEPTIDES order.
    """
    L = len(seq)
    counts = np.zeros(len(TRIPEPTIDES), dtype=float)
    if L < 3:
        return counts

    total = 0
    for i in range(L - 2):
        a1, a2, a3 = seq[i], seq[i+1], seq[i+2]
        if a1 in AA_INDEX and a2 in AA_INDEX and a3 in AA_INDEX:
            tp = a1 + a2 + a3
            idx = TRIPEP_INDEX[tp]
            counts[idx] += 1
            total += 1

    if total > 0:
        counts /= total
    return counts

# Sanity check
test_tri = compute_tripeptide_composition(df["sequence"].iloc[0])
print("3-mer vector length:", len(test_tri))
print("3-mer sum:", test_tri.sum())
```

```python
# =========================
# Cell 4: Feature extraction
# =========================

def extract_features(df):
    """
    Build feature matrix X and label vector y using:
      - AAC (20)
      - Dipeptide composition (400)
      - Sequence length (1)
    Total: 421 features.
    """
    X_list = []
    y_list = []
    ids = []

    for _, row in df.iterrows():
        seq_id = row["id"]
        seq = row["sequence"]
        label = row["label"]

        aac = compute_aac(seq)                      # 20
        dipep = compute_dipeptide_composition(seq)  # 400
        length = np.array([len(seq)], dtype=float)  # 1

        feats = np.concatenate([aac, dipep, length])
        X_list.append(feats)
        y_list.append(label)
        ids.append(seq_id)

    X = np.vstack(X_list)
    y = np.array(y_list, dtype=int)
    ids = np.array(ids)
    return X, y, ids

X, y, ids = extract_features(df)
print("Feature matrix shape:", X.shape)  # (N_samples, 421)
```

```python
# =========================
# Cell A2: extract features with 3-mers
# =========================

def extract_features_with_trimers(df):
    """
    AAC (20) + dipeptide (400) + 3-mers (8000) + length (1)
    Total: 8421 features.
    """
    X_list = []
    y_list = []
    ids = []

    for _, row in df.iterrows():
        seq_id = row["id"]
        seq = row["sequence"]
        label = row["label"]

        aac    = compute_aac(seq)                     # 20
        dipep  = compute_dipeptide_composition(seq)   # 400
        trim   = compute_tripeptide_composition(seq)  # 8000
        length = np.array([len(seq)], dtype=float)    # 1

        feats = np.concatenate([aac, dipep, trim, length])
        X_list.append(feats)
        y_list.append(label)
        ids.append(seq_id)

    X = np.vstack(X_list)
    y = np.array(y_list, dtype=int)
    ids = np.array(ids)
    return X, y, ids

# Build tri-mer feature matrix
X_tri, y_tri, ids_tri = extract_features_with_trimers(df)
print("Tri-mer feature matrix shape:", X_tri.shape)
```

```python
# =========================
# Cell A3: RF with 3-mer features (train/test)
# =========================

X_train_tri, X_test_tri, y_train_tri, y_test_tri = train_test_split(
    X_tri, y_tri, test_size=0.2, stratify=y_tri, random_state=RANDOM_STATE
)

rf_tri = RandomForestClassifier(
    n_estimators=1000,
    max_depth=30,
    min_samples_split=4,
    min_samples_leaf=2,
    class_weight="balanced",
    n_jobs=-1,
    random_state=RANDOM_STATE,
)

rf_tri.fit(X_train_tri, y_train_tri)

y_pred_tri  = rf_tri.predict(X_test_tri)
y_proba_tri = rf_tri.predict_proba(X_test_tri)[:, 1]

print("=== RF with 3-mer features (test set) ===")
print(f"Accuracy:  {accuracy_score(y_test_tri, y_pred_tri):.4f}")
print(f"Precision: {precision_score(y_test_tri, y_pred_tri):.4f}")
print(f"Recall:    {recall_score(y_test_tri, y_pred_tri):.4f}")
print(f"F1-score:  {f1_score(y_test_tri, y_pred_tri):.4f}")
print(f"MCC:       {matthews_corrcoef(y_test_tri, y_pred_tri):.4f}")
print(f"AUC:       {roc_auc_score(y_test_tri, y_proba_tri):.4f}")
```

```python
# ======================================
# Cell 5: Train-test split + RF training
# ======================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=RANDOM_STATE,
)

rf = RandomForestClassifier(
    n_estimators=1000,
    max_depth=30,
    min_samples_split=4,
    min_samples_leaf=2,
    class_weight="balanced",
    n_jobs=-1,
    random_state=RANDOM_STATE,
)

rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)
y_proba = rf.predict_proba(X_test)[:, 1]

acc  = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec  = recall_score(y_test, y_pred)
f1   = f1_score(y_test, y_pred)
mcc  = matthews_corrcoef(y_test, y_pred)
auc  = roc_auc_score(y_test, y_proba)

print("=== Test set performance ===")
print(f"Accuracy:  {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall:    {rec:.4f}")
print(f"F1-score:  {f1:.4f}")
print(f"MCC:       {mcc:.4f}")
print(f"AUC:       {auc:.4f}")
```

```python
# =========================
# Cell A4: 5-fold CV with 3-mer features
# =========================

# Define scoring metrics for cross-validation
scoring = {
    "accuracy": "accuracy",
    "precision": "precision",
    "recall": "recall",
    "f1": "f1",
    "roc_auc": "roc_auc",
}

rf_tri_cv = RandomForestClassifier(
    n_estimators=1000,
    max_depth=30,
    min_samples_split=4,
    min_samples_leaf=2,
    class_weight="balanced",
    n_jobs=-1,
    random_state=RANDOM_STATE,
)

cv_results_tri = cross_validate(
    rf_tri_cv,
    X_tri,
    y_tri,
    cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE),
    scoring=scoring,
    n_jobs=-1,
)

print("=== 5-fold CV (RF + 3-mers) ===")
for metric, values in cv_results_tri.items():
    if metric.startswith("test_"):
        name = metric.replace("test_", "")
        print(f"{name:10s}: mean={values.mean():.4f}, std={values.std():.4f}")
```

```python
# ======================================
# Cell 6: 5-fold cross-validation (RF)
# ======================================

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

scoring = {
    "accuracy": "accuracy",
    "precision": "precision",
    "recall": "recall",
    "f1": "f1",
    "roc_auc": "roc_auc",
}

rf_for_cv = RandomForestClassifier(
    n_estimators=1000,
    max_depth=30,
    min_samples_split=4,
    min_samples_leaf=2,
    class_weight="balanced",
    n_jobs=-1,
    random_state=RANDOM_STATE,
)

cv_results = cross_validate(
    rf_for_cv,
    X,
    y,
    cv=cv,
    scoring=scoring,
    n_jobs=-1,
)

print("=== 5-fold CV performance ===")
for metric, values in cv_results.items():
    if metric.startswith("test_"):
        name = metric.replace("test_", "")
        print(f"{name:10s}: mean={values.mean():.4f}, std={values.std():.4f}")
```

```python
# ======================================
# Cell 7: Feature importance overview
# ======================================

# Names: 20 AAC, 400 dipeptides, 1 length
aac_names = [f"AAC_{aa}" for aa in AA_ORDER]
dipep_names = [f"DIPEP_{dp}" for dp in DIPEPTIDES]
feat_names = aac_names + dipep_names + ["SEQ_LEN"]

importances = rf.feature_importances_
imp_df = pd.DataFrame({
    "feature": feat_names,
    "importance": importances
}).sort_values("importance", ascending=False)

print("Top 20 most important features:")
print(imp_df.head(20))
imp_df.to_csv("feature_importances.csv", index=False)
```

```python
# ======================================
# Cell 8: Predict 5 random sequences
# ======================================

# Predictions for full dataset using trained rf
all_probs = rf.predict_proba(X)[:, 1]
all_preds = rf.predict(X)

results = df.copy()
results["pred_label"] = all_preds
results["pred_proba_RBP"] = all_probs

sample5 = results.sample(n=5, random_state=RANDOM_STATE)

print("=== Predictions for 5 random proteins ===\n")
for _, row in sample5.iterrows():
    print(f"ID:            {row['id']}")
    print(f"True label:    {row['label']}   (1 = RBP, 0 = non-RBP)")
    print(f"Predicted:     {row['pred_label']}")
    print(f"P(RBP | seq):  {row['pred_proba_RBP']:.4f}")
    seq = row["sequence"]
    print(f"Seq (first 60 aa): {seq[:60]}{'...' if len(seq) > 60 else ''}")
    print("-" * 60)
```
